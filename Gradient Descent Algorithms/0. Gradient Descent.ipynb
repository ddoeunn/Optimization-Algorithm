# **Unconstrained Minimization** 
Minimize $f(x)$ where $f: \mathbb{R}^n \rightarrow \mathbb{R}$ is convex and twice continuously differentiable (which implies that dom $f$ is open). Assume there exists an optimal point $x^*$ and denote the optimal value, $\inf_x f(x) = f(x^*) = p^*$. A necessary and sufficient condition for a point $x^*$ to be optimal is $\nabla f(x^*) = 0$, since $f$ is differentiable and convex. Typically use iterative algorithms to get the optimal solution $x^*$. Compute sequence of iterates $x_k$ that (hopefully) converge to $x^*$. $x_{k+1}$ is some (simple) function of $f$, previous iterates.

---
## **Descent Algorithms**  
Pick direction $p_k$ such that $f(x_k + \alpha p_k) < f(x_k)$ for some $\alpha > 0$. There are many choices of such direction $p_k$. (gradient descent / conjugate gradient / newton / ...) Which is the direction that is the "steepest" of them all?  
* By taylor's theorem, $f(x_k + \alpha p) = f(x_k) + \alpha p^T \nabla f_k + \frac{1}{2}\alpha^2 p^T \nabla ^2 f(x_k + tp)p$ for some $t \in (0, \alpha)$  
* Rate of change fo $f$ along direction $p$ : $\lim_{\alpha \rightarrow 0}\frac{f(x_k+\alpha p)-f(x_k)}{\alpha} = p^T \nabla f_k$  
* Unit direction $p$ with most rapid decrease : $min_p p^T \nabla f_k$ ,subject to $\lVert p \rVert = 1$ $\Rightarrow -\nabla f_k / \lVert \nabla f_k \rVert$  

Can we characterize "descent" direction?  
* By Talor's theorem, $f(x_k + \epsilon p_k) = f(x_k) + \epsilon p_k^T \nabla f_k + O(\epsilon^2)$  
* Suppose angle between $p_k$ and $\nabla f_k$ is $\theta_k$, and $cos(\theta_k) < 0$. i.e. angle is strictly less than 90 degrees.  
$\Rightarrow p_k^T \nabla f_k = \lVert p \rVert \lVert \nabla f_k \rVert Cos(\theta_k) < 0 \Rightarrow f(x_k + \epsilon p_k) < f(x_k)$  
* Any downhill direction is a descent direction


## **Gradient Descent**  
### **Algorithm**  
> **given** a starting point $x \in dom(f)$  
**repeat**  
    1.$\Delta x := -\nabla f(x)$  
    2.Line search. Choose step size $t$ via exact or backtracking line search  
    3.Update $x:=x+t\Delta x$  
**until** stopping criterion is satisfied.  
The stopping criterion is usually of the form $\lVert \nabla f(x) \rVert_2 \le \eta$, where $\eta$ is small and positive.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def getLoss(x1, x2):
    return (10 * x1**2 + x2**2) * 0.5

xx1, xx2 = np.meshgrid(np.linspace(-20, 20, 1000), np.linspace(-20, 20, 1000))
x1, x2 = xx1.ravel(), xx2.ravel()
fx = getLoss(x1, x2).reshape(xx1.shape)
```

### **How to choose step size?**

#### **Gradient Descent with Fixed Step Size**

```python
def gradientDescentFixed(start, alpha=.1, iter_max=1000, eps=1e-3):
    x_old = start
    x_history = [x_old]
    loss_history = []
    for i in range(iter_max):
        grad = np.array([10, 1]) * x_old
        x_new = x_old - alpha * grad
        x_history.append(x_new)
        loss = getLoss(*x_new)
        loss_history.append(loss)
        if (loss > 1e+5) | (loss < eps) | (np.abs(np.sum(x_new - x_old)) < eps):
            break
        x_old = x_new
    return x_old, np.array(x_history), loss_history, i
```

```python
start = np.array([2, 20])
case = ['too Big', 'just Right', 'too Small']
alphas = [.3, .1, .001]
plt.figure(figsize=(15, 5))
for i, alpha in enumerate(alphas):
    x_star, x_history, loss, iteration = gradientDescentFixed(start, alpha)
    plt.subplot(1, 3, i+1)
    plt.contour(xx1, xx2, fx, 10)
    plt.plot(x_history[:, 0], x_history[:, 1], lw=.5, marker="o", ms=.5, c='black') 
    plt.scatter(0, 0, marker="*", c='black')
    plt.xlim(-20, 20)
    plt.ylim(-20, 20)
    plt.title(case[i]+', iteration = ' + str(iteration))
```

#### **Gradient Descent with Backtracking Line Search**  
>**given** a descent direction $\Delta x = -\nabla f(x)$ ,$\alpha \in (0, 0.5)$ ,$\beta \in (0, 1)$ ,$t:=1$  
**while** $f(x + t \Delta x) > f(x) + \alpha t \nabla f(x)^T \Delta x$ ,$t:=\beta t$

```python
def gradientDescentBacktracking(start, iter_max=1000, eps=1e-3):
    x_old, alpha, beta = start, .4, .5 
    x_history = [x_old]
    loss_history = []
    for i in range(iter_max):
        t = 1
        grad = np.array([10, 1]) * x_old
        loss = getLoss(x_old[0], x_old[1])
        for j in range(iter_max):
            x_new = x_old - t * grad
            loss_new = getLoss(*x_new)
            if loss_new < (loss - alpha * t * np.sum(grad**2)): break
            t *= beta
        if (loss < eps) | (np.abs(np.sum(x_new - x_old)) < eps): break
        x_history.append(x_new)
        loss_history.append(loss_new)
        x_old = x_new
    return x_old, np.array(x_history), loss_history, i
```

```python
start = np.array([2, 20])
x_star, x_history, loss, iteration = gradientDescentBacktracking(start)
plt.contour(xx1, xx2, fx, 10)
plt.plot(x_history[:, 0], x_history[:, 1], lw=.5, marker="o", ms=.5, c='black') 
plt.scatter(0, 0, marker="*", c='black')
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.title('Backtracking, iteration = ' + str(iteration))
plt.show()
```

**Beale function**  
$f(x, y) = (1.5 - x + xy)^{2} + (2.25 - x + xy^{2})^{2} + (2.625 - x +xy^{3})^{2}$  
global minima : $(x, y) = (3, 0.5)$

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm

def beale(x, y):
    z = (1.5 - x + x * y)**2 + (2.25 - x + x * y**2)**2 + (2.625 - x +
                                                           x * y**3)**2
    return z


def bealeGrad(x, y):
    dx = 2. * ((1.5 - x + x * y) * (y - 1) + (2.25 - x + x * y**2) *
               (y**2 - 1) + (2.625 - x + x * y**3) * (y**3 - 1))
    dy = 2. * ((1.5 - x + x * y) * x + (2.25 - x + x * y**2) * 2. * x * y +
               (2.625 - x + x * y**3) * 3. * x * y**2)
    return dx, dy


def draw_beale():
    xmin, xmax, xstep = -4.5, 4.5, .2
    ymin, ymax, ystep = -4.5, 4.5, .2
    xs_ = np.arange(xmin, xmax, xstep)
    ys_ = np.arange(ymin, ymax, ystep)
    x, y = np.meshgrid(xs_, ys_)
    z = beale(x, y)

    minima = np.array([3., .5])
    minima_ = minima.reshape(-1, 1)
    z_minima = beale(*minima)

    plt.figure(figsize=(8, 5))
    ax = plt.axes(projection='3d', elev=50, azim=-50)
    ax.plot(*minima_, z_minima, 'r*', markersize=10)
    ax.plot_surface(x, y, z, norm=LogNorm(), rstride=1,cstride=1, edgecolor='None',
                    alpha=0.3,cmap=plt.cm.jet)
    ax.view_init(30, 10)
    ax.set_title("Beale Function Visualization")
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')
    ax.set_xlim((xmin, xmax))
    ax.set_ylim((ymin, ymax))
    return ax
```

```python
class GradientDescent():
    iter_max, eps = 1000, 1e-6
    
    def __init__(self, fx, gx, x_init, y_init, learning_rate):
        self.fx, self.gx = fx, gx
        self.x, self.y = x_init, y_init
        self.lr = learning_rate
        self.x_history, self.y_history, self.z_history= [], [], []
        print('x_init = {}, y_init = {}, learning rate = {}'.format(
            self.x, self.y, self.lr))

    def save_history(self, x, y, z):
        self.x_history.append(x)
        self.y_history.append(y)
        self.z_history.append(z)

    def gradientDescent(self):
        z_old = self.fx(self.x, self.y)
        for i in range(self.iter_max):
            if (i) % 100 == 0:
                print('{}th iteration: z = {:.4f}, x = {:.4f}, y = {:.4f}'.format(i, z_old, self.x, self.y))
            self.save_history(self.x, self.y, z_old)
            dx, dy = self.gx(self.x, self.y)
            self.x = self.x - self.lr * dx
            self.y = self.y - self.lr * dy
            z_new = self.fx(self.x, self.y)
            if np.abs(z_old - z_new) < self.eps: break
            z_old = z_new
        print('{}th iteration: z = {:.4f}, x = {:.4f}, y = {:.4f}'.format(i, z_old, self.x, self.y))
        self.x_history = np.array(self.x_history)
        self.y_history = np.array(self.y_history)
        self.z_history = np.array(self.z_history)
    
    def draw_loss(self):
        iteration = self.z_history.shape[0]
        plt.scatter(range(iteration), self.z_history, s=.3)
        plt.title('learning_rate = {}, iteration = {}'.format(self.lr, iteration))
        plt.show()
```

```python
GD = GradientDescent(beale, bealeGrad, .7, 1.4, .01)
GD.gradientDescent()
GD.draw_loss()
ax = draw_beale()
ax.plot(GD.x_history, GD.y_history, GD.z_history, label='Gradient Descent', color='b')
plt.show()
```

```python
GD = GradientDescent(beale, bealeGrad, -2, 2, .001)
GD.gradientDescent()
GD.draw_loss()
ax = draw_beale()
ax.plot(GD.x_history, GD.y_history, GD.z_history, label='Gradient Descent', color='b')
plt.show()
```
