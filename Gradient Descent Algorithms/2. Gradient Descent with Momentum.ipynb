# **Acceleration**  
Gradient Descent relies too much on local information to decide direction and hence might be too slow. With additional 'momentum' term, it might be less slow.  Momentum prevents oscillation due to local-driven (i.e. gradient direction)

## **Gradient Descent with Momentum**  
Apart from moving through the current Gradient, it moves further in that direction, remembering the way it had moved before.  
: $x_{k+1} = x_k \alpha_k \nabla f(x_k) + \beta_k (x_k - x_{k-1})$  
Can be written as a purely descent type method  
\begin{cases}
p_k = -\nabla f(x_k) + \beta_k p_{k-1} \\
x_{k+1} = x_k + \alpha p_k
\end{cases}

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm
```

```python
def beale(x, y):
    z = (1.5 - x + x * y)**2 + (2.25 - x + x * y**2)**2 + (2.625 - x +
                                                           x * y**3)**2
    return z


def bealeGrad(x, y):
    dx = 2. * ((1.5 - x + x * y) * (y - 1) + (2.25 - x + x * y**2) *
               (y**2 - 1) + (2.625 - x + x * y**3) * (y**3 - 1))
    dy = 2. * ((1.5 - x + x * y) * x + (2.25 - x + x * y**2) * 2. * x * y +
               (2.625 - x + x * y**3) * 3. * x * y**2)
    return dx, dy


def draw_beale():
    xmin, xmax, xstep = -4.5, 4.5, .2
    ymin, ymax, ystep = -4.5, 4.5, .2
    xs_ = np.arange(xmin, xmax, xstep)
    ys_ = np.arange(ymin, ymax, ystep)
    x, y = np.meshgrid(xs_, ys_)
    z = beale(x, y)

    minima = np.array([3., .5])
    minima_ = minima.reshape(-1, 1)
    z_minima = beale(*minima)

    plt.figure(figsize=(8, 5))
    ax = plt.axes(projection='3d', elev=50, azim=-50)
    ax.plot(*minima_, z_minima, 'r*', markersize=10)
    ax.plot_surface(x,
                    y,
                    z,
                    norm=LogNorm(),
                    rstride=1,
                    cstride=1,
                    edgecolor='None',
                    alpha=0.3,
                    cmap=plt.cm.jet)
    ax.view_init(30, 10)
    ax.set_title("Beale Function Visualization")
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')
    ax.set_xlim((xmin, xmax))
    ax.set_ylim((ymin, ymax))
    return ax
```

```python
class GradientDescentMomentum():
    iter_max, eps = 1000, 1e-6
    
    def __init__(self, fx, gx, x_init, y_init, momentum, learning_rate):
        self.fx, self.gx = fx, gx
        self.x, self.y, self.v = x_init, y_init, np.zeros(2)
        self.mmt, self.lr = momentum, learning_rate
        self.x_history, self.y_history, self.z_history= [], [], []
        print('x_init = {}, y_init = {}, v_init = {}, momentum = {}, \
              learning rate = {}'.format(self.x, self.y, self.v, self.mmt, self.lr))

    def save_history(self, x, y, z):
        self.x_history.append(x)
        self.y_history.append(y)
        self.z_history.append(z)
    
    def update(self, grad):
        self.v = -grad + self.mmt * self.v
        self.x = self.x + self.lr * self.v[0]
        self.y = self.y + self.lr * self.v[1]
        
    def gradientDescentMomentum(self):
        z_old = self.fx(self.x, self.y)
        for i in range(self.iter_max):
#             if (i) % 100 == 0:
#                 print('{}th iteration: z = {:.4f}, x = {:.4f}, y = {:.4f}'.format(i, z_old, self.x, self.y))
            self.save_history(self.x, self.y, z_old)
            grad = np.array(self.gx(self.x, self.y))
            self.update(grad)
            z_new = self.fx(self.x, self.y)
            if np.abs(z_old - z_new) < self.eps: break
            z_old = z_new
#         print('{}th iteration: z = {:.4f}, x = {:.4f}, y = {:.4f}'.format(i, z_old, self.x, self.y))
        self.x_history = np.array(self.x_history)
        self.y_history = np.array(self.y_history)
        self.z_history = np.array(self.z_history)
    
    def draw_loss(self):
        iteration = self.z_history.shape[0]
        plt.scatter(range(iteration), self.z_history, s=.3)
        plt.title('momentum = {}, learning_rate = {}, iteration = {}'.format(self.mmt, self.lr, iteration))
        plt.show()
```

```python
GDM = GradientDescentMomentum(beale, bealeGrad, .7, 1.4, .9, .01)
GDM.gradientDescentMomentum()
GDM.draw_loss()
ax = draw_beale()
ax.plot(GDM.x_history, GDM.y_history, GDM.z_history, label='Gradient Descent', color='b')
plt.show()
```

```python
GDM = GradientDescentMomentum(beale, bealeGrad, -2, 2, .9, .001)
GDM.gradientDescentMomentum()
GDM.draw_loss()
ax = draw_beale()
ax.plot(GDM.x_history, GDM.y_history, GDM.z_history, label='Gradient Descent', color='b')
plt.show()
```

```python
GDM = GradientDescentMomentum(beale, bealeGrad, -3, -3, .5, .0001)
GDM.gradientDescentMomentum()
GDM.draw_loss()
ax = draw_beale()
ax.plot(GDM.x_history, GDM.y_history, GDM.z_history, label='Gradient Descent', color='b')
plt.show()
```

## **Nestrov Gradient Descent**  

```python
class GradientDescentNestrov():
    iter_max, eps = 1000, 1e-6
    
    def __init__(self, fx, gx, x_init, y_init, momentum, learning_rate):
        self.fx, self.gx = fx, gx
        self.x, self.y, self.v = x_init, y_init, np.zeros(2)
        self.mmt, self.lr = momentum, learning_rate
        self.x_history, self.y_history, self.z_history= [], [], []
        print('x_init = {}, y_init = {}, v_init = {}, momentum = {}, \
              learning rate = {}'.format(self.x, self.y, self.v, self.mmt, self.lr))

    def save_history(self, x, y, z):
        self.x_history.append(x)
        self.y_history.append(y)
        self.z_history.append(z)
    
    def update(self, grad):
        self.v = -self.lr * grad + self.mmt * self.v
        self.x = self.x + self.v[0]
        self.y = self.y + self.v[1]
        
        
    def gradientDescentNestrov(self):
        z_old = self.fx(self.x, self.y)
        for i in range(self.iter_max):
#             if (i) % 100 == 0:
#                 print('{}th iteration: z = {:.4f}, x = {:.4f}, y = {:.4f}'.format(i, z_old, self.x, self.y))
            self.save_history(self.x, self.y, z_old)
            x_next = self.x + self.mmt * self.v[0]
            y_next = self.y + self.mmt * self.v[1]
            grad = np.array(self.gx(x_next, y_next))
            self.update(grad)
            z_new = self.fx(self.x, self.y)
            if np.abs(z_old - z_new) < self.eps: break
            z_old = z_new
#         print('{}th iteration: z = {:.4f}, x = {:.4f}, y = {:.4f}'.format(i, z_old, self.x, self.y))
        self.x_history = np.array(self.x_history)
        self.y_history = np.array(self.y_history)
        self.z_history = np.array(self.z_history)
    
    def draw_loss(self):
        iteration = self.z_history.shape[0]
        plt.scatter(range(iteration), self.z_history, s=.3)
        plt.title('momentum = {}, learning_rate = {}, iteration = {}'.format(self.mmt, self.lr, iteration))
        plt.show()
```

```python
GDN = GradientDescentNestrov(beale, bealeGrad, .7, 1.4, 0.9, .01)
GDN.gradientDescentNestrov()
GDN.draw_loss()
ax = draw_beale()
ax.plot(GDN.x_history, GDN.y_history, GDN.z_history, label='Gradient Descent', color='b')
plt.show()
```

```python
GDN = GradientDescentNestrov(beale, bealeGrad,  -2, 2, .9, .001)
GDN.gradientDescentNestrov()
GDN.draw_loss()
ax = draw_beale()
ax.plot(GDN.x_history, GDN.y_history, GDN.z_history, label='Gradient Descent', color='b')
plt.show()
```

```python
GDN = GradientDescentNestrov(beale, bealeGrad, -3, -3, .5, .0001)
GDN.gradientDescentNestrov()
GDN.draw_loss()
ax = draw_beale()
ax.plot(GDN.x_history, GDN.y_history, GDN.z_history, label='Gradient Descent', color='b')
plt.show()
```
